```python
def copy_stock_charts(tickers):
    source_dir = '/Users/arun/Documents/Python Projects/Floorsheet Code/All scrip accumulation charts'
    dest_dir = '/Users/arun/Documents/Python Projects/Floorsheet Code/Selected stocks'

    # Ensure destination directory exists
    os.makedirs(dest_dir, exist_ok=True)

    # Clear destination directory
    for filename in os.listdir(dest_dir):
        file_path = os.path.join(dest_dir, filename)
        if os.path.isfile(file_path):
            os.remove(file_path)

    # Copy each ticker's PNG file
    for ticker in tickers:
        filename = f"{ticker}.png"
        source_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(dest_dir, filename)

        if os.path.exists(source_path):
            shutil.copy2(source_path, dest_path)
            print(f"Copied '{filename}' to Selected stocks folder.")
        else:
            print(f"File '{filename}' not found in source directory.")
copy_stock_charts(tickers)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

def detect_accumulation_stocks(data: pd.DataFrame, trading_days: int = 30, price_slope_thresh: float = 0.001, vpt_slope_thresh: float = 0.005):
    """
    Detects stocks likely under accumulation using VPT and price slope.

    Parameters:
    - data: DataFrame with ['Date', 'Ticker', 'Close', 'Turnover']
    - trading_days: number of recent trading days to analyze
    - price_slope_thresh: max abs slope of price to consider sideways
    - vpt_slope_thresh: min slope of VPT to be considered rising

    Returns:
    - List of tickers showing signs of accumulation
    """
    data['Date'] = pd.to_datetime(data['Date'])
    data = data.sort_values(['Ticker', 'Date'])
    
    tickers = data['Ticker'].unique()
    accumulation_candidates = []

    for ticker in tickers:
        stock = data[data['Ticker'] == ticker].tail(trading_days).copy()
        if stock.shape[0] < trading_days // 2:
            continue  # not enough data

        # Compute VPT
        stock['DailyReturn'] = stock['Close'].pct_change().fillna(0)
        stock['VPT'] = (stock['Turnover'] * stock['DailyReturn']).cumsum()

        # Prepare regression input: X = time in days, y = Close / VPT
        X = np.arange(len(stock)).reshape(-1, 1)

        price_model = LinearRegression().fit(X, stock['Close'].values.reshape(-1, 1))
        vpt_model = LinearRegression().fit(X, stock['VPT'].values.reshape(-1, 1))

        price_slope = price_model.coef_[0][0]
        vpt_slope = vpt_model.coef_[0][0]

        # Check if price is sideways and VPT is trending up
        if abs(price_slope) <= price_slope_thresh and vpt_slope >= vpt_slope_thresh:
            accumulation_candidates.append(ticker)

    return accumulation_candidates
```

```python
detect_accumulation_stocks(adjusted_price, price_slope_thresh= 0.005, vpt_slope_thresh= 0.009)
```
